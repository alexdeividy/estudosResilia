## Consultas

### Funções iloc e loc
- iloc acessa por posição (linha, coluna)
- loc permite acessar os dados pelo index

```python
walmart.loc[0]['Date']
```

```python
walmart[:1]['Date'] #slice
```

```python
walmart.iloc[0,1] #pesquisa pelo indice da linha e indice da coluna
```

```python
walmart.loc[[1,3,4]][['Weekly_Sales', 'Date']] #para buscar algumas linhas, passar a lista dentro do loc [[lista aqui]]
```

```python
walmart['Holiday_Flag']
```

- .loc também permite criar condições para retornar

```python
walmart.loc[walmart['Holiday_Flag'] == 1]
```

```python
walmart.loc[(walmart['Temperature'] > 90) & (walmart['Store'] == 1)]
```

# Função Query
- É possível realizar consultas por coluna e por index

```python
walmart.query('Weekly_Sales > 1624383.75')
```

```python
walmart.query('1 <= index < 7')
```

```python
val = 1643690.90
walmart.query('Weekly_Sales > @val')
```

```python
walmart.query('Store in [1,6]')
```
